```python
# feature engineering process
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
```

```python
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
final_df.info()
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.info()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute']=final_df['Arrival_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(columns=['Arrival_Time'],axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(15)
```

```python
final_df['Dept_Hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_Minute']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.info()
```

```python
final_df['Dept_Hour']=final_df['Dept_Hour'].astype(int)
final_df['Dept_Minute']=final_df['Dept_Minute'].astype(int)
```

```python
final_df.drop(columns=['Dep_Time'],axis=1,inplace=True)
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop(columns=['Route'],axis=1,inplace=True)
final_df.info()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df[final_df['Duration']=='5m']
# observations
# duration of flight from mum to hyd cannot be 5 minutes so we have to drop those invalid records
```

```python
final_df = final_df[final_df['Duration'] != '5m']
final_df[final_df['Duration']=='5m']
```

```python
# as per me the duration can be calculate by the dept time and arrival time so there is no need of the duration column
# hence dropping it
final_df.drop(columns=['Duration'],axis=1,inplace=True)
final_df.info()
```

```python
final_df['Airline'].unique()
# trying label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
from sklearn.preprocessing import OneHotEncoder # onehot encoding
ohe=OneHotEncoder(sparse_output=False)
final_df['Airline']=ohe.fit_transform(final_df[['Airline']])
final_df['Source']=ohe.fit_transform(final_df[['Source']])
final_df['Destination']=ohe.fit_transform(final_df[['Destination']])
final_df['Additional_Info']=ohe.fit_transform(final_df[['Additional_Info']])
```
